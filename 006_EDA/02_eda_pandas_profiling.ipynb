```python
import pandas as pd
from ydata_profiling import ProfileReport

# 1. Load your dataset
df = pd.read_csv(r'../005_featured_data/gurgaon_properties_featured.csv').drop_duplicates()

# 2. Use minimal=True to bypass the SciPy bug
# This still provides the full summary of variables/columns
profile = ProfileReport(
    df,
    title="Full Report",
    explorative=True,
    correlations={
        "pearson": {"calculate": False},
        "spearman": {"calculate": False},
        "kendall": {"calculate": False},
        "cramers": {"calculate": False},
        "phi_k": {"calculate": False},
    }
)

# 3. Generate the report
profile.to_file("output_report.html")

print("Report successfully created.")
```
